```python

def rad_to_degree():
  radian=float(input("enter a value : "))
  degree=radian*57.29
  print("radian to degree:",degree)
def degree_to_radian():
  degree=float(input("enter a value :")) 
  radian=degree*0.01745
  print("Radian to degree:",radian) 
def num_fac():
  num=int(input("enter number: "))  
  fac=1
  for i in range(1,num+1):
    fac=fac*i
    
  print("factorial is :",fac)
def int_float():
  b=int(input("enter integer value : "))
  a=float(b)
  print("float value is :",a)
def float_int():
  b=float(input("enter float value : "))
  
  print("integer value is : ",int(b))
def ms_to_kmhr():
  m=float(input("enter value in m/s :"))  
  k=(m*36)/10
  print("speed in km/hr is :",k)
def kmhr_to_ms():
  k=float(input("enter value in km/hr :"))
  m=(k*10)/36
  print("speed in m/s is : ",m)  
def char_to_ascii():
  k=input("enter charechter : ")
  m=ord(k)
  print("Ascii value of entered charechter is : ",m)
def ascii_to_char():
  k=int(input("enter ascii value:"))
  m=chr(k) 
  print("Charechter is :",m) 
def cel_to_far():
  c=float(input("enter temp in celsius: ")) 
  f=(c*1.8)+32
  print("temperature in farenhite :",f) 
def far_to_cel():
  f=float(input("enter temp in farenhite :"))
  c=(f-32)*0.555
  print("temperature in celsius :",c)  
def  num_sqrt():
  a=float(input("enter number "))
  b=a**0.5
  print("square root of the number :",b)
def mega_nibb():
  m=float(input("enter value in megabytes :"))  
  n=2000000*m
  print("Value in Nibble is : ",n)
def frac_to_dec():
  a=int(input("enter numerator:"))
  b=int(input("enter denominator :"))
  
  c=float(a/b)

  
  print("decimal number is : ",c)
def hex_dec():
  x=input("Enter Hexa Value")
  dec=int(x,16)
  print("Hexadecimal Value ",x)
  print("Decimal Value",dec)
def dec_hexa():
  x=int(input("Enter decimal Value : ")) 
  i=0
  hexdecnum=[]    
  while x!=0:
    rem=x%16
    if rem<10:
      rem=rem+48
    else:
      rem=rem+55
    rem=chr(rem)  
    hexdecnum.insert(i,rem) 
    i=i+1
    x=int(x/16) 
  print("Hexadecimal Value is :") 
  i=i-1
  while i>=0:
    print(end=hexdecnum[i]) 
    i=i-1
  print()  
def oct_to_dec():
  octnum=int(input("enter octal number : "))  
  i=0
  chk=0
  decnum=0
  while octnum!=0:
    rem=octnum%10
    if rem>7:
      chk=1
      break
    decnum=decnum+(rem*(8**i))  
    i=i+1
    octnum=int(octnum/10)
  if chk==0:
    print("Decimal Number is",decnum)  
  else:
    print("Invalid Input")  
def dec_to_oct():
  decnum=int(input("Enter decimal number "))
  i=0
  octnum=[]
  while decnum!=0:
    rem=decnum%8
    octnum.insert(i,rem)
    i=i+1
    decnum=int(decnum/8)
  print("Equivalent Octal Number is :")  
  i=i-1
  while i>=0:
    print(octnum[i],end="")
    i=i-1
  print()  
def dec_to_bin():
  d=int(input("Enter decimal number:"))
  i=0
  b=[]
  while d!=0:
    rem=d%2
    b.insert(i,rem)
    i=i+1
    d=int(d/2) 
  i=i-1
  print("Binary Value is : ")
  while i>=0:
    print(end=str(b[i]))
    i=i-1
  print()  
def bin_to_dec():
  b=int(input("enter binary number:")) 
  d=0
  i=1
  while b!=0:
    rem=b%10
    d=d+(rem*i)
    i=i*2
    b=int(b/10)
  print("Decimal number is :",d)  



  
      


  

  


menu={}
menu['1']="Radian to Degree"
menu['2']="Degree to Radian"
menu['3']="Number to Factorial"
menu['4']="Integer To Float"
menu['5']="Float To INTEGER"
menu['6']="M/s To KM/hr"
menu['7']="M/s to Km/Hr"
menu['8']="Charechter To ASCII"
menu['9']="Ascii to Charechter"
menu['10']="Celsius To Farenhite"
menu['11']="Farenhite To Celsius"
menu['12']="Number to squareroot"
menu['13']="Mega bytes to Nibbles"
menu['14']="Fraction to Decimal"
menu['15']="HexaDecimal To Decimal"
menu['16']="Decimal To HexaDecimal"
menu['17']="octal to decimal"
menu['18']="Decimal To Octal"
menu['19']="Decimal to Binary"
menu['20']="Binary To Decimal"
menu['21']="Exit"
while True :
  options=menu.keys()
  for i in options:
    print(i,menu[i])
  o=input("enter which operation to perform:") 
  if o=='1':
    print("Radian To Degree")
    rad_to_degree()
  elif o=='2':
    print("Degree To Radian")
    degree_to_radian()
  elif o=='3':
    print('Number To Factorial')
    num_fac()
  elif o=='4':
    print("Integer To Float")
    int_float()
  elif o=='5':
    print("Float To Integer")
    float_int()    
  elif o=='6':
    print("M/S To KM/HR")
    ms_to_kmhr()  
  elif o=='7':
    print("KM/HR TO M/S")
    kmhr_to_ms()
  elif o=='8':
    print("Charechter To Ascii Value")
    char_to_ascii()
  elif o=='9':
    print("Ascii value to Charechter")
    ascii_to_char()      
  elif o=='10':
    print("Celsius to Farenhite")
    cel_to_far()
  elif o=='11':
    print("Farenhite to Celsius")
    far_to_cel()  
  elif o=='12':
    print("Number to SquareRoot")
    num_sqrt()
  elif o=='13':
    print("MegabytesTo Nibbles")
    mega_nibb()   
  elif o=='14':
    print("Fraction To Decimal")
    frac_to_dec()   
  elif o=='15':
    print('Hexadecimal To Decimal')
    hex_dec()
  elif o=='16':
    print("Decimal To HexaDecimal")
    dec_hexa()
  elif o=='17':
    print("Octal To Decimal") 
    oct_to_dec()  
  elif o=='18':
    print("Decimal to Octal")  
    dec_to_oct()  
  elif o=='19':
    print("Decimal To Binary")
    dec_to_bin()  
  elif o=='20':
    print("Binary To Decimal")
    bin_to_dec()  
  elif o>='21':
    break
  else:
    print("Invalid Input")  







  
  

      

    
    
```
